## Descenso del Gradiente 

```python
# Importar librerias
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

Funci√≥n:

$f(x,y) = x^2 + y ^2$

```python
f = lambda X: X[0]**2 + X[1]**2
df = lambda X: [2*X[0], 2*X[1]]

# Prepara los datos para dibuar mapa de niveles de Z

resolucion = 30
rango = 2.5
X = np.linspace(-rango, rango, resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion, resolucion))

for ix, x in enumerate(Y):
    for iy, y in enumerate(Y):
        Z[iy, ix] = f([x, y])

contour = plt.contourf(X, Y, Z, resolucion)
cbar = plt.colorbar(contour)

# Generar punto de partida aleatorio
P = [random.uniform(-rango, rango), 
    random.uniform(-rango, rango)]
plt.plot(P[0], P[1], "o", c = "red")
TA = 0.05

for _ in range(500):
    grad = df(P)
    P[0], P[1] = P[0] - TA*grad[0], P[1] - TA*grad[1]
    plt.plot(P[0], P[1], "o", c = "red")

plt.plot(P[0], P[1], "o", c = "green")
print("Solution: ", P, f(P))
```

Funci√≥n:

$$
f(x) = \sin(\frac{1}{2}\cdot x^2 - \frac{1}{4}\cdot y^2 + 3)\cdot \cos(2\cdot x + 1 - e^y)
$$

```python
# Definimos la funci√≥n
# sin(1/2 * x^2 - 1/4 * y^2 + 3) * cos(2*x + 1 - e^y)

f = lambda X: np.sin(1/2 * X[0]**2 - 1/4 * X[1]**2 + 3) * np.cos(2*X[0] + 1 - np.e**X[1])
```

```python
# Aproximamos el valor del gradiente en un punto por su funci√≥n
 
def df(PUNTO, h = 0.01):
    T = np.copy(PUNTO)
    grad = np.zeros(2)
    for it, th in enumerate(PUNTO):
        T[it] = T[it] + h
        grad[it] = (f(T) - f(PUNTO))/h
    return grad
```

```python
resolucion = 30
rango = 2.5
X = np.linspace(-rango, rango, resolucion)
Y = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion, resolucion))

for ix, x in enumerate(Y):
    for iy, y in enumerate(Y):
        Z[iy, ix] = f([x, y])

contour = plt.contourf(X, Y, Z, resolucion)
cbar = plt.colorbar(contour)

# Generar punto de partida aleatorio
Pi = [random.uniform(-rango, rango), 
    random.uniform(-rango, rango)]
plt.plot(Pi[0], Pi[1], "o", c = "white")

# Valor de la taza de aprendizaje
TAi = 0.05

# Se genera el primer punto del descenso de gradiente
Pj = np.zeros(2)
Pk = np.zeros(2)
Pj[0], Pj[1] = df(Pi, TAi)

# Distancia euclidiana entre el punto aleatorio y el 
# primer punto generado
dist_ec = np.linalg.norm(Pi-Pj)

for _ in range(1000):
    # Taza de aprendizaje dentro del bucle
    TA = TAi
    # Segundo punto generado del descenso de gradiente (punto n+1)
    Pk[0], Pk[1] = df(Pj, TA)
    # Distancia euclidea entre el punto n y el punto n+1
    dist_ec2 = np.linalg.norm(Pk-Pj)
    
    '''Si la distancia euclidea n+1 es menor que la distancia
    euclidea n, entonces la taza de aprendizaje cambia con 
    base en el cocientre entre ambas distancias euclideas.
    Caso contrario se conserva el valor anterior de la taza 
    de aprendizaje.
    '''
    if dist_ec2 < dist_ec:
        TA = (dist_ec2/dist_ec)*TAi
    else:
        TA = TAi
    
    '''Se actualiza los valores de la distancia euclidea y
    el punto n'''
    
    dist_ec = dist_ec2
    Pj[0], Pj[1] = Pk[0], Pk[1]
    
    # Se dibuja el punto n
    plt.plot(Pj[0], Pj[1], "o", c = "red")

plt.plot(P[0], P[1], "o", c = "green")
print("Solution: ", P, f(P))
```
